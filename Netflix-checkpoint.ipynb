<h3 style="color: #556B2F"> Promedio de los mejores top 10, en Netflix desde abril del 2020 </h3

```python
dfFinal[["Days In Top 10"]]
```

```python
bins=np.linspace(min(dfFinal["Days In Top 10"]), max (dfFinal["Days In Top 10"]), 4)
bins
```

```python
group_names=["Malo", "Bueno", "Exc"]
dfFinal["Days In Top 10-rangos"]=pd.cut(dfFinal["Days In Top 10"], bins, labels=group_names, include_lowest=True)
dfFinal[["Days In Top 10", "Days In Top 10-rangos"]].head(5)
```

```python
dfFinal["Days In Top 10-rangos"].value_counts()
```

```python
dfFinal.head(3)
```

```python
gfl=dfFinal.groupby(["Days In Top 10-rangos"]).count()
gfl
```

```python
gfl= gfl.iloc[:,:1]
gfl
```

```python
eje_x=("Malo", "Bueno", "Exc")
eje_y=gfl["As of"].tolist()

##crear grafica con cambio de color, y suavizar el garfico con alpha
plt.bar(eje_x, eje_y, color="#778899", alpha=0.5)

#leyenda en el eje y
plt.ylabel("Promedio según el top 10")

#leyenda en el eje x
plt.xlabel("Rango")

#Titulo
plt.title("Rango de shows, según los días en el Top 10")

#Tamaño del grafico
plt.figure(figsize=(10,30))



plt.show()
```
